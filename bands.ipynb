```python
from pymatgen.ext.matproj import MPRester
from pymatgen.entries.compatibility import MaterialsProjectCompatibility
from pymatgen.electronic_structure.bandstructure import BandStructure
from pymatgen.electronic_structure.plotter import BSPlotter,BoltztrapPlotter,BSPlotterProjected

with MPRester("j4TaClKeSNgPZb8Bxm") as m:
    
#Bandstructure for material id
    bandstructure = m.get_bandstructure_by_material_id("mp-753098")

#Analyse

    #Bande interdite

    bandint = bandstructure.get_band_gap()
    print(bandint)
    dirbandint = bandstructure.get_direct_band_gap()
    print(dirbandint)
    
    #Direction dernière bande de valence
    
    min = BandStructure.get_vbm(bandstructure)
    print("min = ", min['energy'])
    
    #Direction première bande de conduction
    
    max = BandStructure.get_cbm(bandstructure)
    print("max = ", max['energy'])


    #Figure des bandes électroniques du matériau
    
    fig = BSPlotter(bandstructure)
    fig.get_plot(True, None, False, True, None)
    fig1 = BSPlotterProjected(bandstructure)
    fig1.get_elt_projected_plots()
```
